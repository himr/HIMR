
## 【链表】

    实现单链表、双链表、循环链表，支持增删改操作
    实现单链表反转
    实现两个有序链表合并为一个有序链表
    实现求链表的中间结点

```python
# 单链表

class Node():
    """define the node of linklist"""

    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next


class LinkList():
    """define class of linklist"""

    def __init__(self):
        self.head = Node()
        self.length = 0

    def init_linklist(self, data):
        """initial linklist"""

        p = self.head
        for i in data:
            node = Node(i)
            p.next = node
            p = p.next
            self.length += 1

    def isEmpty(self):
        """Determine if the linklist is empty"""
        return self.length == 0

    def insert(self, _index, data):
        """Insert an element into the linklist"""

        if _index < 1 or _index > self.length+1:
            print('IndexError: out of index, fail to insert')
            return
        p = self.head
        node = Node(data)
        count = 0
        while count < _index-1:
            p = p.next
            count += 1
        node.next = p.next
        p.next = node
        self.length += 1

    def delt(self, _index):
        """Delete the value of the element corresponding to _index"""

        if _index < 1 or _index > self.length:
            print('IndexError: out of index, fail to delt')
            return
        p = self.head
        count = 0
        while count < _index-1:
            p = p.next
            count += 1
        p.next = p.next.next
        self.length -= 1

    def node_updata(self, _index, data):
        """Replace the value of the element corresponding to _index with data"""

        if _index < 1 or _index > self.length:
            print('IndexError: out of index, fail to updata')
            return
        p = self.head.next
        count = 1
        while count < _index:
            p = p.next
            count += 1
        p.data = data

    def getItem(self, _index):
        """Get the value of the element corresponding to _index"""

        if _index < 1 or _index > self.length:
            print('IndexError: out of index, fail to getitem')
            return
        p = self.head.next
        count = 1
        while count < _index:
            p = p.next
            count += 1
        return p.data

    def getIndex(self, data):
        """Get the index of the element data"""

        if self.isEmpty():
            print('the linklist is empyt, fail to getindex')
            return
        p = self.head.next
        count = 1
        while p and p.data != data:
            p = p.next
            count += 1
        if count == self.length+1:
            print('%s not found' % str(data))
            return
        else:
            return count

    def traverse_linklist(self):
        """Traversing the linklist"""

        p = self.head.next
        if p == None:
            print('empty linklist')   
        while p:
            print(p.data, end=' ')
            p = p.next

    def reverse_linklist(self):       
        pre = None
        _next = None
        p = self.head.next
        while p:
            _next = p.next
            p.next = pre
            pre = p
            p = _next
        self.head.next = pre

        p = self.head.next
        while p:
            yield p.data
            p = p.next
```

```python
# 双链表

class Node:
    """define the node of linklist"""

    def __init__(self, data=None, prior=None, next=None):
        self.data = data
        self.prior_item = prior
        self.next_item = next


def init_linklist(self, data):
    """initial linklist"""
    p = self.head
    for i in data:
        node = Node(i, prior=p)
        p.next_item = node
        p = p.next_item
        self.length += 1

def insert(self, _index, data):
    """Insert an element into the linklist"""
    if _index < 1 or _index > self.length+1:
        print('IndexError: out of index, fail to insert')
        return
    p = self.head
    node = Node(data)
    count = 0
    while count < _index-1:
        p = p.next_item
        count += 1
    node.next_item = p.next_item
    p.next_item = node
    node.prior_item = p
    if node.next_item != None:
        node.next_item.prior_item = node
    self.length += 1

def delt(self, _index):
    """Delete the value of the element corresponding to _index"""
    if _index < 1 or _index > self.length:
        print('IndexError: out of index, fail to delete')
        return
    p = self.head
    count = 0
    while count < _index-1:
        p = p.next_item
        count += 1
    p.next_item = p.next_item.next_item
    p.next_item.prior_item = p
    self.length -= 1
```
