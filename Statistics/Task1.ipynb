# 统计学基础

## E12 样本与总体
    
    在统计学中，样本的均值通常用，总体的均值用。均值，中位数，众数是用来衡量数据集中趋势。均值：mean 衡量集中趋势的方法，中位数：median 考虑集中趋势的另一种方式，众数：mode 也就是出现频率最高的数字。极差：range数字越大表示数据之间越分散、越小表示越紧密，用来衡量集中趋势。中程数：midrange 衡量集中趋势的另一种方式。
    
## E13 总体方差

    总体方差是用来描述样本数据离散程度的统计工具。即数据偏离样本中间水平的程度，用符号σ^2来表示。
    
## E14 样本方差

    　样本方差是用来描述样本数据离散程度的统计工具。即数据偏离样本中间水平的程度，用符号S^2来表示
    
## E15 标准差

    标准差就是方差的开方，用σ表示。意义与方差一样，优点是直观。
    
    
## E16 诸方差

    诸方差是方差的简化，样本平方的均值减去平均数的平方。
    
## E17 随机变量

    随机变量通常为函数，用于量化随机过程，通常用X,Y,Z等表示，普通变量为小写的x,y,z表示。随机变量分为离散型和连续型。
    
## E18 概率密度函数

    概率密度函数的公式为
   $$ f(x) = \frac{1}{\sigma\sqrt{2\pi}}e^-\frac{1}{2}(\frac{x-\mu}{\sigma})^2 $$ 
   $$ 当\mu = 0,\sigma=1 时 \ f(x)= \frac{1}{\sqrt{2\pi}}e^(-\frac{x^2}{2}) 为标准正态分布。 $$


## E18-22 二项分布

    二项分布就是伯努利实验重复N次。每次实验结果都有两个可能，并且两个结果相互独立，与实验的次数无关，也就是实验的事件概率每次都保持独立不变。
    
## E23 期望

    随机变量的期望值即为总体的均值。当总体样本无限多，均值计算很难得到结果，提出期望E（X）计算均值的方法，利用频率作为权重计算所有结果的加权平均。
    
## E24 二项分布的期望值

    二项分布X~B(n,p),期望值E(X)=np。
    
## E25\26 泊松过程

    泊松过程来自二项分布，它是求取某个时间段内发生事件X概率有多大，并且是离散分布的。

# 正态分布


## E27 大数定律

    大数定律描述了随机现象最根本的一个性质：平均结果的稳定性。大数定律告诉我们：对于独立同分布的随机序列，只要总体均值（随机变量期望）存在，则随着样本数的增加，样本均值会收敛到总体均值。注意样本数的足够性，概率是频率的一个极限值，这样可以避免赌徒谬误。
    
## E28-34 正态分布

      二项分布，泊松分布都是离散分布，而正态分布是连续分布。二项分布和泊松分布都可以转化为正态分布。泊松分布是 n*p = λ,而正态 n*p 分布是为无穷大。正态分布的数值一般与均值相差一个标准差概率是68%， 两个标准差概率是95%，三个标准差概率是99.7%。
