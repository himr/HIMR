    标签为：td class = "postbody"

## 1.2 获取内容

```python
import urllib
import requests
from bs4 import BeautifulSoup
import json
```

```python
xyz = requests.Session()
url = 'http://www.dxy.cn/bbs/thread/626626#626626'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/60.0.3100.0 Safari/537.36'}
conditions = requests.Request(method='GET',url=url,headers=headers)
prepared = dxy.prepare_request(conditions)
res = dxy.send(prepared)
html = res.text
soup = BeautifulSoup(html,'lxml')
```

```python
soup
```

```python
contents = dict()
```

```python
def get_contents(html):
    for i in html.find_all('tbody'):
        try:
            name = i.find("div", class_ = "auth").get_text(strip = True)
            content = i.find("td", class_ = "postbody").get_text(strip=True)
            contents[name] = content
        except:
            pass
```

```python
get_contents(soup)
```
